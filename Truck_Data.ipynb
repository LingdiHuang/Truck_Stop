```python
def converse(x):
    for k,v in us_state_abbrev.items():
        if x==v:
            return k 
```

```python
df['State'] = df.apply(lambda x: converse(x.St),axis=1)
```

```python
df.rename(columns = {'Co':'County'},inplace=True)
```

```python
df[df.State.isnull()]
```

```python
df.State.fillna('Alabama',inplace=True)
```

```python
s = df[df.apply(lambda x: x['Rte'].startswith('I'),axis=1)]
s.shape
```

```python
def ExtName(rawString):
    RoadName = re.findall(r'I-\d+',rawString)
    if RoadName:        
        for i in RoadName:
            if len(i)<=4:
                return i
            else:
                return None
    else:
        RoadName = re.findall(r'I\d+',rawString)
        for j in RoadName:
            s = j[0]+'-'+j[1:]
            if len(s)<=4:
                return s
            else:
                return None
```

```python
df['Interstate'] = df.apply(lambda x:ExtName(x.Rte),axis=1)
```

```python
df.Interstate.unique()
```

```python
df['if_inter'] = df.apply(lambda x: 0 if x.Interstate is None else 1,axis = 1)
```
